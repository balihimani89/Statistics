Compute probabilities of the following:

\begin{equation}
P(Cap-shape \cap Cap-surface\cap Cap-color | Class)\\[10pt]
P(Cap-shape \cap Cap-surface\cap Cap-color | Class = p)\\[10pt]
\end{equation}

Find the probablity of equation first
\begin{equation}
P(Cap-shape \cap Cap-surface\cap Cap-color | Class = e)\\[10pt]
\end{equation}


Let 
\begin{equation}
P(Cap-shape=x
  \cap Cap-surface=y\cap Cap-color=w| Class = e)=P(A|B)\\[10pt]
\end{equation}




So accordingly the definition for conditional probability we have
Let 
\begin{equation}
P(A|B)=\frac{P(A\cap B)}{P(B)}\\[10pt]
\end{equation}



\begin{equation}
P(A\cap B)=P(Cap-shape=x
  \cap Cap-surface=y\cap Cap-color=w\cap Class=e)\\[10pt]
\end{equation}

```python
data[(data['cap-shape']=="x") & (data['cap-surface']=="y") & (data['cap-color']=="w") & (data['class']=="e")]
```

```python
frequency=data[(data['cap-shape']=="x") & (data['cap-surface']=="y") & (data['cap-color']=="w") & (data['class']=="e")].shape[0]
```

```python
frequency
```

```python
total_mushrooms=data.shape[0]
total_mushrooms
```

```python
numerator_joint_probability=frequency/total_mushrooms
numerator_joint_probability
```

```python
frequency_e=data[data['class']=="e"].shape[0]
frequency_e
```

```python
denominator_e=frequency_e/total_mushrooms
denominator_e
```

```python
conditional_probability=numerator_joint_probability/denominator_e
conditional_probability
```

```python
frequency/frequency_e
```

Lets now first compute probability for 
\begin{equation}
P(Cap-shape \cap Cap-surface\cap Cap-color | Class)\\[10pt]
\end{equation}

```python
conditional_probabilities={}
for clas in data['class'].unique():
    for cap_shape in data['cap-shape'].unique():
        for cap_surface in data['cap-surface'].unique():
            for cap_color in data['cap-color'].unique():
                conditional_prob=(data[(data['cap-shape']==cap_shape) & (data['cap-surface']==cap_surface) & (data['cap-color']==cap_color) & (data['class']==clas)].shape[0])/(data[data['class']==clas].shape[0])
                key_string="P(cap-shape={}".format(cap_shape) + ",cap-surface={}".format(cap_surface) +",cap-color={}".format(cap_color)+"|class={})".format(clas)
                conditional_probabilities[key_string]=conditional_prob
```

```python
conditional_probabilities
```

Now lets compute for 
\begin{equation}
P(Cap-shape \cap Cap-surface\cap Cap-color | Class = e)\\[10pt]
\end{equation}

```python
data_class_p=data[data['class']=="p"]
data_class_p
```

```python
conditional_probabilities_p={}
for clas in data_class_p['class'].unique():
    for cap_shape in data_class_p['cap-shape'].unique():
        for cap_surface in data_class_p['cap-surface'].unique():
            for cap_color in data_class_p['cap-color'].unique():
                conditional_prob_p=(data_class_p[(data_class_p['cap-shape']==cap_shape) & (data_class_p['cap-surface']==cap_surface) & (data_class_p['cap-color']==cap_color) & (data_class_p['class']==clas)].shape[0])/(data_class_p[data['class']==clas].shape[0])
                key_string_p="P(cap-shape={}".format(cap_shape) + ",cap-surface={}".format(cap_surface) +",cap-color={}".format(cap_color)+"|class=p)"
                conditional_probabilities_p[key_string_p]=conditional_prob_p
```
