# compute the probability that P(MATH)>=50,P(SCIENCE)>=50,P(PHYSICS)>=50 on a condition that P(HINDI)>=50.

```python
import numpy as np
```

```python
marks_of_students=np.random.randint(low=0,high=100,size=(1000,4))
marks_of_students
```

```python
import pandas as pd
```

```python
list_of_marks_of_students=pd.DataFrame(data=marks_of_students,columns=('Math','Science','physics','hindi'))
```

```python
list_of_marks_of_students
```

```python
hindi_above_50=list_of_marks_of_students[list_of_marks_of_students['hindi']>=50]
```

```python
boolean_mask=(hindi_above_50['Math']>=50) & (hindi_above_50['Science']>=50) & (hindi_above_50['physics']>=50)
```

```python
above_50_in_Math_Science_physics_hindi=hindi_above_50[boolean_mask]
above_50_in_Math_Science_physics_hindi
```

```python
P_PHYSICS_MATH_SCIENCE_ON_HINDI=above_50_in_Math_Science_physics_hindi.shape[0]/hindi_above_50.shape[0]
P_PHYSICS_MATH_SCIENCE_ON_HINDI
```
